```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 



print(f'''
{'='*50}
Program ini akan menyelesaikan data - data dari 
   Percobaan Gerak Benda Menggelinding (M9)
{'='*50 }''')

print(f'''{'-'*50}
1. Percobaan Momen Inersia Silinder Pejal
{'-'*50}''')

# Data dari Silinder
m = 0.412
d = 0.08
r = d/2
g = 9.8

# Data Praktikum
teta = np.array([5,10,15,20,25,30])
a = np.array([0.601,1.155,1.699,1.934,2.829,3.386])
sinteta = np.sin(np.radians(teta))
I = (m*(r**2)/a)*(g*np.sin(np.radians(teta))-a)
n = len(a)

# Jumlah (Ip - Irata2)**2
Irata = sum(I)/n
IR= sum((I-Irata)**2)

# Nilai ralat2 nyaa
RM = (IR/(n-1))**(1/2)
RN = (RM/Irata)*100

# Nilai Inersia Teori 
Ite = 1/2*(m*(r**2))
Rel = (abs(Ite-Irata)/Ite)*100


# Plotting ke dalam tabel 
print(f'''Dengan 
massa silinder = {m} kg , 
jari - jari silinder = {r} m 
''')
data = {
    'teta':teta,
    'sin teta':sinteta,
    'Percepatan':a,
    'Momen Iner':I,
    '(Ipr - I rata2)^2':(I-(sum(I)/len(teta))**2)
}

pd.set_option('display.colheader_justify', 'center')
df = pd.DataFrame(data)
df_center = df.style.set_properties(**{'text-align':'center'})
display(df_center)
print(f'''{'-'*50}
Rata - rata I praktikum = {sum(I)/len(teta):.3e} kg m2
Jumlah (Ip - Irata2)^2  = {IR:.3e} kg m2 
Ralat Mutlak (RM)       = {RM:.3e} 
Ralat Nisbi  (RN)       = {RN:.3f} %
{'-'*50}
Kesalahan Relatif (%)   = {Rel:.3f} %
''')
print(f'''{'-'*50}
2. Percobaan Momen Inersia Silinder Berongga 
{'-'*50}''')
# Data dari Silinder
m = 0.499
ddal = 0.075
dlua = 0.065
rdal = ddal/2
rlua = dlua/2
g = 9.8

# Data Praktikum
teta = np.array([5,10,15,20,25,30])
a = np.array([0.470,0.957,1.371,2.001,2.666,2.689])
sinteta = np.sin(np.radians(teta))
I = (m*(rdal**2+rlua**2)/a)*(g*np.sin(np.radians(teta))-a)
n = len(a)

# Jumlah (Ip - Irata2)**2
Irata = sum(I)/n
IR= sum((I-Irata)**2)

# Nilai ralat2 nyaa
RM = (IR/(n-1))**(1/2)
RN = (RM/Irata)*100

# Nilai Inersia Teori 
Ite = 1/2*(m*((rdal**2)+(rlua**2)))
Rel = (abs(Ite-Irata)/Ite)*100


# Plotting ke dalam tabel 
print(f'''Dengan 
massa silinder = {m} kg , 
jari - jari dalam = {rdal} m, 
jari - jari luar = {rlua}m
''')
data = {
    'teta':teta,
    'sin teta':sinteta,
    'Percepatan':a,
    'Momen Iner':I,
    '(Ipr - I rata2)^2':(I-(sum(I)/len(teta))**2)
}

df = pd.DataFrame(data)
df_center = df.style.set_properties(**{'text-align':'center'})
display(df_center)
print(f'''{'-'*50}
Rata - rata I praktikum = {sum(I)/len(teta):.3e} kg m2
Jumlah (Ip - Irata2)^2  = {IR:.3e} kg m2 
Ralat Mutlak (RM)       = {RM:.3e} 
Ralat Nisbi  (RN)       = {RN:.3f} %
{'-'*50}
Momen Inersia Teori     = {Ite:.3e} kg m2 
Kesalahan Relatif (%)   = {Rel:.3f} %
''')

```
