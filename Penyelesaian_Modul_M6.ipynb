```python
import numpy as np
import pandas as pd
```

```python
n = 5
# Keadaan Awal benda A dalam Keadaan diam
va = np.zeros(n)
ta = np.zeros(n)
ma = np.zeros(n)
pa = np.zeros(n)
for i in range (n):
  ma[i] = 140.5

# Keadaan awal Benda B
mb = np.array([140.5,153,165.5,178,190.5])
tb = np.array(([0.173,0.176,0.176,0.186,0.157]))
vb = 3/tb
pb = mb*vb

# Keadaan Setelah tumbukan
t_a = np.array([0.185,0.180,0.173,0.177,0.149])
v_a = 3 / t_a
p_a = ma*v_a

# Keadaan benda B Setelah Tumbukan adalah Diam 
t_b = np.zeros(n)
v_b = np.zeros(n)
p_b = np.zeros(n)



data = {
    'mA' : ma,
    'tA' : ta,
    'pA' : pa,
    'mB' : mb,
    'tB' : tb,
    'vB' : vb,
    'pB' : pb,
    "t'A":t_a,
    "v'A":v_a,
    "p'A":p_a,
    "t'B":t_b,
    "v'B":v_b,
    "p'B":p_b,
}
print('='*90,'\n                        Tabel 1 Data Pengukuran Tumbukan elastis ')
print('-'*90)
pd.set_option('display.float_format', lambda x : f"{x:.3f}")
df = pd.DataFrame(data)
display(df)
if sum(pb) == sum(p_a):
  a =('Hukum Kekekalan Momentum Berlaku pada Bercobaan ini')
else :
  a =('Hukum Kekekalan Momentukm tidak Berlaku ')
print(f'''{"="*90}
Apakah Hukum Kekekalan Momentum berlaku pada percobaan ini ??
Jawabannya adalah {a}
P sebelum = P sesudah 
{sum(pa)+sum(pb):.3e} = {sum(p_a)+sum(p_b):.3e} Tidak sama antara kedua nilainya 
{"="*90}''')
```

```python
n = 5
# Keadaan Awal benda A dalam Keadaan diam
va = np.zeros(n)
ta = np.zeros(n)
ma = np.zeros(n)
pa = np.zeros(n)
for i in range (n):
  ma[i] = 140.5

# Keadaan awal Benda B
mb = np.array([140.5,153,165.5,178,190.5])
tb = np.array(([0.02244,0.03080,0.02479,0.03551,0.03675]))
vb = 3/tb
pb = mb*vb

# Keadaan Setelah tumbukan
t_ab = np.array([0.0974,0.238,0.0834,0.113,0.127])
v_ab = 3 / t_ab
p_ab = (ma+mb)*v_ab




data = {
    'mA' : ma,
    'tA' : ta,
    'pA' : pa,
    'mB' : mb,
    'tB' : tb,
    'vB' : vb,
    'pB' : pb,
    "t'AB":t_ab,
    "v'AB":v_ab,
    "p'AB":p_ab
}
print('='*90,'\n                      Tabel 1 Data Pengukuran Tumbukan Tidak Elastis ')
print('-'*90)
pd.set_option('display.float_format', lambda x : f"{x:.4f}")
df = pd.DataFrame(data)
display(df)
if sum(pb) == sum(p_ab):
  a =('Hukum Kekekalan Momentum Berlaku pada Bercobaan ini')
else :
  a =('Hukum Kekekalan Momentukm tidak Berlaku ')
print(f'''{"="*90}
Apakah Hukum Kekekalan Momentum berlaku pada percobaan ini ??
Jawabannya adalah {a}
P sebelum = P sesudah 
{sum(pa)+sum(pb):.3e} = {sum(p_ab):.3e} Tidak sama antara kedua nilainya 
{"="*90}''')
```

```python
# Bandingakan Nilai Energi Kinetik Sebelum dan Sesudah 
def Ek(p,v): 
  return 0.5 * p * v 

# Energi Kinetik Sebelum 
Ek1 = Ek(pb,vb)

# Energi Kinetik Sesudah 
Ek2 = Ek(p_ab,v_ab)

data = {
    'Energi Kinetik Sebelum ':Ek1,
    'Energi Kinetik Sesudah ':Ek2
}

pd.set_option('display.float_format', lambda x : f'{x:.3e}')
df = pd.DataFrame(data,index=[1,2,3,4,5])
display(df)
print(f'{"="*50}\nJumlah    {sum(Ek1):16.3e}{sum(Ek2):24.3e}')
print(f'''{"="*50}''')
```
